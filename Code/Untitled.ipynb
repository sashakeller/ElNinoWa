```python
# Luke: define system-specific file paths here 
datapath  =   '../Data/Raw/'
filename  =   'sst.mnmean.nc'

#import xarray module for python to load netcdf data
import xarray as xr
import numpy as np

#define nino3.4 region 
lat_bounds=[5, -5]
lon_bounds=[190, 240]

#load data
ds = xr.open_dataset(datapath + filename)

#pull out sst variable
sst = ds['sst']

# Subset lat/lon data over regio want to average
sst_nino34 = sst.sel(lat=slice(lat_bounds[0], lat_bounds[1]),
                         lon=slice(lon_bounds[0], lon_bounds[1]))

#weight the data by latitude grid box area
w = np.cos(np.deg2rad(sst_nino34.lat))
w = w.broadcast_like(sst_nino34)

#create spatial average in Nino3.4 region using weights
sst_nino34_mean = (sst_nino34 * w).mean(dim=['lat', 'lon']) / w.mean(dim=['lat', 'lon'])

#select last 30 years, remove the average
sst_nino34_mean_1992_2021 = sst_nino34_mean.sel(time=slice('1992','2021'))
sst_nino34_mean_1992_2021_anom = sst_nino34_mean_1992_2021 - sst_nino34_mean_1992_2021.mean(dim='time')
#calculate 3-month rolling average
sst_nino34_mean_1992_2021_anom_rolling = sst_nino34_mean_1992_2021_anom.rolling(time=3, center=True).mean().dropna("time")
#pick which time period want to print out date/time and average SST:
timeslice_nino34 = sst_nino34_mean_1992_2021_anom.sel(time=slice('2000','2021')) 

for ssti,sst in enumerate(timeslice_nino34):
    print(timeslice_nino34.time[ssti].values, sst.values)


```

```python
import pandas as pd 

months=timeslice_nino34
sst=sst.values

Month_Measure=pd.DataFrame([months],[sst])
pd.DataFrame(Month_Measure).to_csv(datapath+"monthly_enso.csv")
```
